```python
""" Question-1:

Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both

"""
#Sol:
def calculate_depth(preorder):
    def helper(preorder, index):
        if index >= len(preorder):
            return 0
        
        if preorder[index] == 'l':  # Leaf node
            return 0
        
        left_index = index + 1
        right_index = index + 2
        
        left_depth = helper(preorder, left_index)
        right_depth = helper(preorder, right_index)
        
        return 1 + max(left_depth, right_depth)
    
    return helper(preorder, 0)

# Example usage:
preorder = "nlnll"
depth = calculate_depth(preorder)
print(depth)
```

```python
""" Question-2:

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.
"""
#Sol:
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def print_left_view(root):
    if root is None:
        return
    
    # Create a dictionary to store the leftmost node at each level
    leftmost_nodes = {}
    
    def dfs(node, level):
        if node is None:
            return
        
        # If the level is not already in the dictionary, add the current node
        if level not in leftmost_nodes:
            leftmost_nodes[level] = node.data
        
        # Traverse the left subtree
        dfs(node.left, level + 1)
        
        # Traverse the right subtree
        dfs(node.right, level + 1)
    
    # Start the DFS traversal from the root at level 1
    dfs(root, 1)
    
    # Print the leftmost nodes at each level
    for level in sorted(leftmost_nodes.keys()):
        print(leftmost_nodes[level], end=' ')

# Example usage:
# Create a binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Print the left view of the binary tree
print_left_view(root)
```

```python
""" Question-3:Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side. """
#Sol:
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def print_right_view(root):
    if root is None:
        return
    
    # Create a dictionary to store the rightmost node at each level
    rightmost_nodes = {}
    
    def dfs(node, level):
        if node is None:
            return
        
        # Update the rightmost node at the current level
        rightmost_nodes[level] = node.data
        
        # Traverse the left subtree
        dfs(node.left, level + 1)
        
        # Traverse the right subtree
        dfs(node.right, level + 1)
    
    # Start the DFS traversal from the root at level 1
    dfs(root, 1)
    
    # Print the rightmost nodes at each level
    for level in sorted(rightmost_nodes.keys()):
        print(rightmost_nodes[level], end=' ')

# Example usage:
# Create a binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.right = Node(4)
root.right.left = Node(5)
root.right.right = Node(6)

# Print the right view of the binary tree
print_right_view(root)

```

```python
""" Question-4 Given a Binary Tree, The task is to print the bottom view from left to right. A node x is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1. """
#Sol:
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.horizontal_distance = 0

def print_bottom_view(root):
    if root is None:
        return
    
    # Create a dictionary to store the bottommost node at each horizontal distance
    bottom_nodes = {}
    
    # Perform a level-order traversal using a queue
    queue = deque([(root, 0)])
    
    while queue:
        node, horizontal_distance = queue.popleft()
        
        # Update the bottom node at the current horizontal distance
        bottom_nodes[horizontal_distance] = node.data
        
        # Update the horizontal distance of the children nodes
        if node.left:
            node.left.horizontal_distance = horizontal_distance - 1
            queue.append((node.left, node.left.horizontal_distance))
        
        if node.right:
            node.right.horizontal_distance = horizontal_distance + 1
            queue.append((node.right, node.right.horizontal_distance))
    
    # Print the bottom nodes from left to right
    for horizontal_distance in sorted(bottom_nodes.keys()):
        print(bottom_nodes[horizontal_distance], end=' ')

# Create the binary tree
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

# Print the bottom view of the binary tree
print_bottom_view(root)
```
